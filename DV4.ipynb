```
import numpy as np
import matplotlib.pyplot as plt
import requests
import urllib.request
import time
from collections import Counter
```

```
from bs4 import BeautifulSoup
url="https://en.wikipedia.org/wiki/Machine_learning"
response = requests.get(url).text
```

```
soup = BeautifulSoup(response, 'lxml')
li=[]
for para in soup.find_all('p'):
  li.append(para.text.split())
```

```
total=0
for i in range(len(li)):
  total+=len(li[i])
print(total)

```

```
lis=[]
for i in li :
  for j in i : 
     lis.append(j)
  
len(lis)  
  
  
      
```

```
lt=[]
for i in lis:
   if lis.count(i) >= 2 :
           lt.append(i)

lt              
```

```
dict={}
for i in lt :
  dict[i]=lis.count(i)
```

```
dict
```

```
val=[]
valc=[]
for i in dict:
 val.append(dict[i])
 valc.append(i)
```

```
val
valc
```

```
plt.pie(val,labels=valc)
```

```
plt.bar(val,valc)
```
